Import tensorflow and print version

```python
import tensorflow as tf
print(tf.__version__)
```

```python
# Create a tensor
scaler = tf.constant(7)
scaler
```

```python
scaler.ndim
```

```python
# Create a vector tensor
vector = tf.constant([1,7])
vector
```

```python
vector.ndim
```

```python
# Create a matrix
matrix = tf.constant([[1,7],[7,8]])
matrix
```

```python
# Create a tensor
tensor = tf.constant([[[1.0,3],[7,8]],
                      [[3,4],[6,3]],[[4,6],[5,8]] ])
tensor
```

```python
tensor.ndim
```
