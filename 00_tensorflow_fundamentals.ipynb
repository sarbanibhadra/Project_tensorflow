### Shuffle a tensor



```python
unshuffle_tensor = tf.constant([[7,5],
                                [6,9],
                                [8,3]])
unshuffle_tensor
```

```python
tf.random.set_seed(5) #global level random seed
tf.random.shuffle(unshuffle_tensor, seed=5) # operation level random seed
```

# Getting information from Tensors
* shape: Number of elements in each dimension
* rank: Number of dimensions
* axis or dimension: Each dimension
* size: Total number of elements


```python
rank_4_tensor = tf.zeros(shape=(2,3,4,5), dtype=tf.int32)
rank_4_tensor
```

```python
# shape
rank_4_tensor.shape
```

```python
# rank
rank_4_tensor.ndim
```

# Find tensorflow variance and standar deviation

```python
import tensorflow_probability as tfp
```

```python
tf.__version__
```

```python
E = tf.random.Generator.from_seed(6)
E = tf.abs(tf.cast(10*E.normal(shape=(6,8)), dtype=tf.int32))
tf.reduce_max(E)
tfp.stats.variance(E)
```

Finding variance using tf.math

```python
tf.math.reduce_variance(tf.cast(E, dtype=tf.float32))
```
